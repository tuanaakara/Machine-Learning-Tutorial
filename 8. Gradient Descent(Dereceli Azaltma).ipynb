## Gradient Descent 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
#%matplotlib inline -->> grafiği kod ile aynı kod bloğunda çizdirmeye yarıyor. farklı kod bloğunda çizdirmiyor.
```

bu fonksiyonun gradyan inişini bulacağız:  f(x) = x3 - 3x2 + 7

```python
#fonksiyonun oluşturulması ve grafiğinin çizdirilmesi
function = lambda x: (x ** 3) - (3 * (x ** 2))+7

# -1 ile 3 arasında eşit aralıklı 1000 sayı üretin (dik bir eğri sağlamak için keyfi olarak seçilmiş bir sayı ...)
x = np.linspace(-1, 3, 300)
#x = np.linspace(-1, 30, 3000) bu şekilde değerler değiştirip bakabiliriz

#plot the curve
plt.plot(x, function(x))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#x'in türevini aldık, hesaplama için fonksiyon yazalım.
def deriv(x): 
    x_deriv = 3* (x**2) - (6 * (x))
    return x_deriv

deriv(20)
```

```python
#Gerçek gradyan inişini adım adım gerçekleştiren bir fonksiyon
def step(x_new, x_prev, precision, l_r):
    
    #her yineleme sırasında güncellenmiş x ve y değerlerininin değiştirileceği boş listeler oluşturma
    x_list, y_list = [x_new], [function(x_new)]
    #istediğiniz eşik değere kadar döngüye devam edin.
    while abs(x_new - x_prev) > precision:
        #cx in değerini değiştir
        x_prev=x_new
        #xin eski değerinin türevini al
        d_x = deriv(x_prev)
        #öncekini, türevin çarpımını ve öğrenme oranını ekleyerek yeni x değerinizi elde edin
        x_new = x_prev - (l_r * d_x)
        #sonucun daha sonra görselleştirilmesi için x'in yeni değerini tüm x'lerin listesine ekleyin
        x_list.append(x_new)
        #sonucun daha sonra görselleştirilmesi için tüm y'lerin listesine y'nin yeni değerini ekleyin
        y_list.append(function(x_new))
    print("Lokal minimum noktası: " + str(x_new))
    print("Adım sayısı: " + str(len(x_list)))

    plt.subplot(1,2,2)
    plt.scatter(x_list, y_list, c="g")
    plt.plot(x_list, y_list, c="g")
    plt.plot(x, function(x), c="r")
    plt.title("Dereceli Azaltma")
    plt.show()

    plt.subplot(1,2,1)
    plt.scatter(x_list, y_list, c="g")
    plt.plot(x_list, y_list, c="g")
    plt.plot(x, function(x), c="r")
    plt.xlim([1.8, 2.5])
    plt.title("Anahtar nokta kısmında büyütülmüş alan")
    plt.show()
```

```python
step(0.8, 0, 0.0001, 0.2)
#step(0.8, 2, 0.0001, 0.4) yaptı hoca ve çok uzun sürdü. bi de #x = np.linspace(-1, 30, 3000) yapmıştı.
```
