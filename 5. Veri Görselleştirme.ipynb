5. uygulamamızda Matplotlib ile Veri Görselleştirme yapacağız.

Çizgi Grafiği, Çubuk Grafikler, Pasta Grafik, Histogram, Dağılım Grafiği ve subplots kullanımlarını göreceğiz.

```python
import matplotlib.pyplot as plt
```

```python
#veri seti
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

#çizgi grafiği oluşturma
plt.plot(x, y)

#başlık ve etiketler
plt.title('Basit Çizgi Grafiği')
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')

#arka planı beyaz yerine grid(ızgara) yapmak
plt.grid()
#plt.grid(axis = 'x')
#plt.grid(axis = 'y')

#Gösterim
plt.show()
```

## Çubuk Grafikler (Bar Plots)

```python
import numpy as np
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 5, 12]
values2 = [4, 6, 8, 10]
values3 = [12, 10, 9, 10]

x = np.arange(len(categories))
print(x)

#plt.bar(x, height, width, bottom, align)
plt.bar(x + 0.2, values1, width= 0.2, label='Set 1')
plt.bar(x, values2, width= 0.2, label='Set 2')
plt.bar(x - 0.2, values3, width= 0.2, label='Set 3')

plt.title('Gruplandırılmış Çubuk Grafiği') #loc title konumu verir.
plt.xlabel('Kategoriler')
plt.ylabel('Değerler')

plt.legend()
plt.show()
```

```python
plt.bar(categories, values1, label='Set 1')
plt.bar(categories, values2, bottom=values1, label='Set 2')

plt.title('Yığılmış Çubuk Grafiği') 
plt.xlabel('Kategoriler')
plt.ylabel('Değerler')

plt.legend()
plt.show()
```

## Pasta Grafikleri (Pie Charts)

```python
labels= ['Elma', 'Muz', 'Kiraz', 'Portakal']
sizes = [15,30,45,10]

plt.pie(sizes, labels=labels, autopct = '%1.1f%%')  #autopct sayısal değerleri graf üzerine yazdırmak için.

plt.title('Meyve Dağılımı')

plt.show()
```

```python
#veri seti
data = np.random.randn(1000)

#histogram oluşturma
plt.hist(data, bins=30) #bins aralıktaki eşit genişlikteki bölmelerin sayısını tanımlar.

#başlık ve etiketler
plt.title('Histogram') 
plt.xlabel('Değerler')
plt.ylabel('Frekans')

#gösterim
plt.show()
```

## Dağılım Grafikleri

```python
x = [1,2,3,4,5]
y = [10,20,25,30,35]

plt.scatter(x, y, marker='*', linewidth=8, color='red')    #linewidth markerın boyutudur.

plt.title('Dağılım Grafiği') 
plt.xlabel('X Değeri')
plt.ylabel('Y Değeri')

plt.show()
```

```python
colors=[10,20,30,40,50]
sizes=[100,200,300,400,500]

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis') 
plt.colorbar() #renk skalası ekler

plt.title('Özelleştirilmiş Dağılım Grafiği') 
plt.xlabel('X Değeri')
plt.ylabel('Y Değeri')

plt.show()
```

```python
plt.figure(figsize=(10,6), dpi=100)
plt.plot(x, y)
plt.show()

#dpi = dots per inch: Grafiklerin çözünürlüğünü 100 dpi olarak ayarlar. Bu, grafiğin her inçte 100 nokta içerdiği anlamına gelir.
```

# Alt Grafikler (Subplots)

```python
fig, axs = plt.subplots(2,2)
fig.suptitle('Dört şekil için ortak başlık')

axs[0,0].plot(x,y)
axs[0,0].set_title('Grafik 1')

axs[0,1].bar(categories,values)
axs[0,1].set_title('Grafik 2')

axs[1,0].hist(data)
axs[1,0].set_title('Grafik 3')

axs[1,1].pie(sizes, labels=sizes, autopct='%1.1f%%')
axs[1,1].set_title('Grafik 4')

plt.tight_layout()   
plt.show()
```

```python
fig, (axs1, axs2) = plt.subplots(2,1,figsize=(6,10),sharex=True)
fig.suptitle('İki Şekil için Ortak Başlık')

axs1.plot(x,y)
axs1.set_title('Grafik 1')

axs2.bar(categories,values)
axs2.set_title('Grafik 2')

plt.tight_layout()
plt.show()

#plt.subplots(2, 1, ...) : 2 satır ve 1 sütundan oluşan bir alt grafik düzeni oluşturur. 2 dikey subplot
#figsize=(6, 10) : figürün genişliği 6 inç, yüksekliği ise 10 inç olarak ayarlanmıştır.
#sharex=True : tüm alt grafiklerin x eksenini paylaşmasını sağlar
#fig, (axs1, axs2) : (axs1, axs2) değişkenleri ise alt grafiklerin (subplots) eksenlerini temsil eder.
```

```python
x = [1,2,3,4]
y = [10,15,40,25]

plt.subplot(2,1,1)
plt.plot(x,y)

x=np.array([1,2,3,4])
y=np.array([10,15,40,25])

plt.subplot(2,1,2)
plt.plot(x,y)

plt.tight_layout()
plt.show()

#Figür, iki subplot grafikten oluşur: bir üstte ve bir altta. Üst subplot grafik plt.subplot(2, 1, 1) ile, alttaki ise plt.subplot(2, 1, 2) ile çizilir.
#plt.tight_layout() fonksiyonu, alt grafiklerin arasındaki boşlukları düzenler ve her şeyin düzgün görünmesini sağlar.
```

```python
import pandas as pd
df = pd.read_csv("car_sales.csv")
df.head()
```

```python
df['Price'] = df['Price'].str.replace('$','').str.replace(',','').astype(float).astype(int) 
```

```python
df
```

```python
df['Sale Date'] = pd.date_range(start ='1-1-2024', periods = len(df))
#date_range-> başlangıç tarihi verdik, df uzunluğumuzu aldı ve son satıra kadar 1 gün artırarak ilerledi ve sale date değişkenimizi doldurdu.
```

```python
df
```

```python
df['Total Sale'] = df['Price'].cumsum()
df
```

```python
df.plot(x = "Sale Date", y = "Total Sale"); 

plt.title('Tarihlere Göre Toplam Satış Grafiği')
plt.xlabel('Tarih')
plt.ylabel('Toplam Satış')
```

```python
#bu ve alttaki kod blokları aynı grafiği çizdirir. 
plt.scatter(df['Odometer (KM)'], df['Price'], marker='*', linewidth=2, color='red')
plt.title('KM Değerine Göre Fiyat Grafiği') 
plt.xlabel('KM')
plt.ylabel('Price')
plt.show()
```

```python
#bu ve üstteki kod blokları aynı grafiği çizdirir. 
df.plot(x = 'Odometer (KM)', y = 'Price', kind = 'scatter')
plt.title('KM Değerine Göre Fiyat Grafiği') 
plt.xlabel('KM')
plt.ylabel('Price')
plt.show()
```
